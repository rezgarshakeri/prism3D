```julia
using Plots
using LinearAlgebra
using Jacobi
gr()
Plots.GRBackend()
```

```julia
function Dubiner(x,y,i,j)
    eta1 = 2*(1 .+ x) ./ (1 .- y) .- 1
    eta2 = y
    D    = jacobi.(eta1, i, 0, 0)
    D   *= (0.5*(1 .- eta2)) .^i
    D   *= jacobi.(eta2, j, 2*i+1, 0)
    return D
end
```

```julia
N = 60
x = LinRange(-1,1,N)
y = LinRange(-1,1,N)
i = 0
j = 1
function f(x,y)
    return x <= -y + 1/(10*N) ? Dubiner(x,y,i,j) : NaN
end
contourf(x, y, f)
#plot(x, y, f, st = :surface, camera = (45,70))
#plot(p[1])
```

```julia
function PrimeBasis(x,y)
    k = 2
    P = []
    for d = 0:k
        for i = 0:d
            push!(P,Dubiner(x,y,i,d-i))
        end
    end
    return P
end
            
```
